# DATA CLEANING

```python
data1.isnull().sum()#representing null values present in data
```

```python
# Drop the column "company" as it has more than 94% data missing
data1 = data1.drop(['company'],axis=1)
```

```python
#replacing null values presnt in data

data1['agent'].fillna(0,inplace=True)

data1['country'].fillna('other',inplace=True)

data1['children'].fillna(data1.children.mean(),inplace=True)#replace null values with mean
```

```python
# Ensure that all the null values get replace and there is no null values in any column
data1.isnull().sum()
```

```python
# Converting datatype of columns 'children', 'company' and 'agent' from float to int.
data1[['children', 'agent']] = data1[['children',  'agent']].astype('int64')
```

```python
#no.of rows of duplicate rows except first based on all columns
duplicates=data1[data1.duplicated()].shape
duplicates
```

```python
#remove duplicate values if any
data1.drop_duplicates(inplace=True)
```

```python


# Adding total people num as column, i.e. total people num = num of adults + children + babies
data1['total_people'] = data1['adults']+data1['children']+data1['babies']
```
